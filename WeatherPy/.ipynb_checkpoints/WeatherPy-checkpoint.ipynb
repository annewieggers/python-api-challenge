```python
# Dependencies
import matplotlib.pyplot as plt
import json
import requests
import pandas as pd
import numpy as np
from api_keys import weather_api_key
from scipy.stats import linregress
from citipy import citipy
from datetime import datetime
import time
import csv
```

## Generate Cities List

```python
# Range of lats and longs
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#List for holding lat_lngs and cities
lat_lngs = []
cities = []
```

```python
# Create a set of random lat and lng combinations and setting lowest and highest
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information and build partial query URL and request results in Fahrenheit
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# Append city from cities list to url
city_url = url + "&q=" + cities[0]

# Make request and store response and JSON-ify response
response = requests.get(city_url).json()

# Print response
print(json.dumps(response, indent=4, sort_keys=True))

```

```python
# create a small list for testing purposes so we don't have to go over +600 cities each time per API call
#cities_small = cities[0:15]
```

```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# Create lists to store data we get from API call 
max_temp = []
city_name = []
countries = []
dates = []
lat = []
lng = []
humidity = []
cloud = []
wind_speed = []

# Counter to go through data
record = 0
sets = 1

for city in cities:
    try: 
        # append first city from list to url
        city_url = url + "&q=" + city
    
        # make the API call
        response = requests.get(city_url).json()
        
        # append data to lists
        city_name.append(response['name'])
        countries.append(response['sys']['country'])
        dates.append(response['dt'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        
        # update count and set number after going through 50 records
        record = record + 1
        if record == 50:
            sets = sets + 1
        
        print(f"Processing Record {record}  of Set {sets} | {city}.")
        
    except:
        print(f"City not found...skipping {city}.")
    
    
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a data frame from responses
weather_dict = {
    "City": city_name,
    "Latitude": lat,
    "Longitude": lng,
    "Max temperature": max_temp,
    "Humidity": humidity,
    "Cloudiness": cloud,
    "Wind speed": wind_speed,
    "Country": countries,
    "Date": dates
}
weather_data = pd.DataFrame(weather_dict)
len(weather_data)
```

```python
# Import data into csv file
weather_data.to_csv("output_data/weather_output.csv", encoding="utf-8", index=False)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#  Create dataframes for northern and southern hempisphere
northern = weather_data.loc[weather_data['Latitude'].astype(float) > 0, :]
southern = weather_data.loc[weather_data['Latitude'].astype(float) < 0, :]

x_values_n = northern['Latitude'].astype(float)
y_values_n = northern['Max temperature'].astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values_n, y_values_n)
regress_values = x_values_n * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f"Regression line equation is: {line_eq}")
print(f"The r-value is: {rvalue}")
plt.scatter(x_values_n,y_values_n, marker="o", alpha=0.4)
plt.plot(x_values_n,regress_values,"r-")
plt.annotate(line_eq,(10,-20),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title("Northern Hemisphere - City Latitude vs. Max Temperature")

plt.show()

# Save the figure
plt.savefig("output_data/North - City Latitude vs. Temperature - Linear regression.png")
```

```python
print("This plot shoes a negative correlation between latitude of cities in the Northern hempisphere (latitude between 0 and 90 degrees) and maximum temperature. The higher the latitude, the lower the temperature. This can be substantiated by the fact that it's very cold (low temperature) on the North Pole (latitude of 90 degrees). ")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression
